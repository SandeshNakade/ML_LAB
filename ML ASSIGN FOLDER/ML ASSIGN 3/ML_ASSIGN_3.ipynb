```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
import seaborn as sns;



data = {
    'student_id': [1,2,3,4,5,6,7,8],
    'age': [18,19,18,20,19,18,20,19],
    'gender': ['Male','Female','Male','Female','Male','Female','Male','Female'],
    'study_hours': [2,4,6,1,5,3,7,2],
    'attendance': [75,82,90,65,88,78,92,70],
    'exam_score': [60,72,85,50,80,68,90,58]
}

df = pd.DataFrame(data)
df
```

```python
df.groupby('gender')['exam_score'].mean().plot(kind='bar')
plt.xlabel('Gender')
plt.ylabel('Average Exam Score')
plt.title('Average Exam Score by Gender')
plt.show()
```

```python
plt.scatter(df['study_hours'], df['exam_score'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score')
plt.show()
```

```python
df_sorted = df.sort_values('study_hours')

plt.plot(df_sorted['study_hours'], df_sorted['exam_score'])
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.title('Study Hours vs Exam Score (Line)')
plt.show()
```

```python
plt.hist(df['exam_score'], bins=5)
plt.xlabel('Exam Score')
plt.ylabel('Frequency')
plt.title('Exam Score Distribution')
plt.show()
```

```python
sns.boxplot(x='gender', y='exam_score', data=df)
plt.title('Exam Score by Gender')
plt.show()
```
