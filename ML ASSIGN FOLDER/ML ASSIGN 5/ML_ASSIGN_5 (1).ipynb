```python
import pandas as pd;
import numpy as np;
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
df = pd.DataFrame({
    'x':[245,312,279,308,199,219,405,324,319,355],
    'y':[1400,1600,1700,1875,1100,1550,2350,2450,1425,1700]
})
```

```python
df.head()
```

```python


sns.boxplot(y=df['y'])
plt.title('Box Plot of House Price')
plt.ylabel('House Price')
plt.show()
```

```python
sns.boxplot(y=df['x'])
plt.title('Box Plot of Square Feet')
plt.ylabel('Square Feet')
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Box Plot of Square Feet and House Price')
plt.show()
```

```python
plt.scatter(df['x'], df['y'])
plt.xlabel('Square Feet')
plt.ylabel('House Price')
plt.title('Square Feet vs House Price')
plt.show()
```

```python


corr = df.corr()
print('Correlation Matrix:')
print(corr)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
Q1 = df['y'].quantile(0.25)
Q3 = df['y'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_y = df[(df['y'] < lower_bound) | (df['y'] > upper_bound)]
outliers_y
```

```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import numpy as np


X = df[['x']]
y = df['y']

model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(X)
```

```python
df['predicted_y'] = model.predict(X)
print(df[['x', 'y', 'predicted_y']])
```

```python
plt.scatter(df['x'], df['y'], label='Actual Data')
plt.plot(df['x'], y_pred, color='red', label='Regression Line')
plt.xlabel('Square Feet')
plt.ylabel('House Price')
plt.title('Linear Regression: Square Feet vs House Price')
plt.legend()
plt.show()
```

```python
print("Slope (m):", model.coef_[0])
print("Intercept (c):", model.intercept_)
```

```python
print("R-squared value:", model.score(X, y))
```

```python


plt.scatter(df['x'], df['y'], label='Actual')
plt.scatter(df['x'], df['predicted_y'], color='red', label='Predicted')
plt.xlabel('Square Feet')
plt.ylabel('House Price')
plt.title('Actual vs Predicted House Prices')
plt.legend()
plt.show()
```
