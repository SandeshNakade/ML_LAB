```python
#USER DEFINED FUNCTION
def mean_score(df_csv):
  total = 0
  count =0
  for score in df_csv["exam_score"]:
      count+=1
      total += score

  return total/count

def min_score(df_csv):
  min = 101
  for score in df_csv["exam_score"]:
    if score < min:
      min = score
  return min

def max_score(df_csv):
  max = 0;
  for score in df_csv["exam_score"]:
    if score > max:
      max = score
  return max

import math

def standard_deviation(df_csv):
    mean = mean_score(df_csv)
    total = 0
    n = len(df_csv)

    for score in df_csv["exam_score"]:
        total += (score - mean) ** 2

    variance = total / n
    std_dev = math.sqrt(variance)

    return std_dev


print(max_score(df_csv))
print(min_score(df_csv))
print(mean_score(df_csv))
print(standard_deviation(df_csv))
```

```python

Q1 = df_csv["exam_score"].quantile(0.25)

Q2 = df_csv["exam_score"].quantile(0.5)
Q3 = df_csv["exam_score"].quantile(0.75)

print("Q1 : ",Q1,"Q2 : ",Q2,"Q3 : ",Q3)

q1 = np.percentile(df_csv["exam_score"],25);

q2 = np.percentile(df_csv["exam_score"],50)
q3 = np.percentile(df_csv["exam_score"],75)

print("q1 : ",q1,"q2 : ",q2,"q3 : ",q3)
```

```python
#OUTLIERS USING QUARTILE
Q1 = df_csv["exam_score"].quantile(0.25)
Q2 = df_csv["exam_score"].quantile(0.5)
Q3 = df_csv["exam_score"].quantile(0.75)

IQR  = Q3 - Q1

lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR

print("LOWER : ",lower)
print("UPPER : ",upper)

outliers = df_csv[(df_csv["exam_score"]<lower) | (df_csv["exam_score"]>upper)]

print(outliers["exam_score"])
```
